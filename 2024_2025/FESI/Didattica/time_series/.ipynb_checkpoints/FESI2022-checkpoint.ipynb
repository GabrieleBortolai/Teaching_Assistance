```python
import os
from scipy.io import loadmat
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter,filtfilt, lfilter
```

```python
patient = 'PT12_40'
data = loadmat(patient+'.mat') #qui metti il path al file .mat sul tuo pc
data_filter = data[patient]

data_o = loadmat(patient+'_original.mat')
data_orig = data_o['data_orig']
```

```python
print(np.shape(data_filter))
print(np.shape(data_orig))
```

```python
# Gaussian filter
from scipy import signal
from scipy.ndimage import convolve1d

inx_data = 0 #try and change matrix column to see other coordinates

windowg = signal.windows.gaussian(30, std=1) #cambiare valori della gaussian
filt=convolve1d(data_orig[:,inx_data],windowg) 
```

```python
plt.figure()
plt.subplot(211)
plt.plot(data_orig[:50,inx_data])
plt.title('original')
plt.subplot(212)
#plt.plot(data_filter[:50,0])
plt.plot(filt[:50])
plt.title('filtered')
#plt.legend(['original','filtered'])
plt.show()
```

```python
fs = 25

fft_f = np.fft.fft(data_orig[:,inx_data])
n = len(fft_f)
freq = np.fft.fftfreq(n, 1/fs)

plt.plot(freq,np.abs(fft_f))  
plt.ylim([-1,100000])
```

```python
idx = np.argwhere(np.abs(freq)>5) #change frequency threshold
fft_cut=np.copy(fft_f)
fft_cut[idx]=0

plt.plot(freq,np.abs(fft_cut) )
plt.xlabel('f [Hz]')
plt.ylim([-1,100000])

plt.grid()
```

```python
f_cut=np.fft.ifft(fft_cut)

plt.figure()
plt.subplot(211)
plt.plot(data_orig[:50,inx_data])
plt.title('original')
plt.subplot(212)

plt.plot(np.real(f_cut[:50]))
plt.title('filtered')
#plt.legend(['original','filtered'])
plt.show()
```
